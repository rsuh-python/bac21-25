## Функции. Рекурсия

Снова немного вернемся к истории и парадигмам программирования. (Ну мы же гуманитарии, нам сам боженька велел понимать философию!)

Итак, для маленьких коротких программ парадигма структурного программирования была достаточной, чтобы сделать их ясными и читаемыми, но программы все росли, и сегодняшние проекты могут достигать тысяч строк кода. Тут уже простая структура не спасет! Нужно прибегать к более сложным методам организации кода.

Для этого в современном мире программирования обычно различают два противоположных подхода:

- функциональное программирование
- объектно-ориентированное программирование

В каком-то смысле это соотносится с категориями "глагол" и "существительное" в естественном языке.

И тот, и другой взгляд на вещи помогают объединять части кода в какие-то категории, но по разным признакам.

Функциональное программирование - это "глагольный" подход: мир состоит из списка действий, которые можно выполнять, а субъекты и объекты не важны. Так, в нашем мире есть функция "приготовить обед", а кто будет в ней участвовать - неважно, может и котик обед готовить, разница будет только в результате.

Объектно-ориентированное программирование - это "существительный" подход: мир состоит из объектов, у которых есть разные атрибуты и характеристики. Объект "котик" умеет мурлыкать и мяукать, это его характеристики.

Функция - это такой объект, который состоит из заранее собранного кода. То есть, допустим, если нам нужно какой-то набор действий выполнять несколько раз в программе, мы можем этот набор вынести за скобочки, сохранить под переменной и вызывать каждый раз, как только нам это понадобится.

Функция - вызываемый объект.

(Питон первоначально задумывался как объектно-ориентированный ЯП, поэтому, хотя он позволяет нам использовать функциональную парадигму, у него все равно все внутри представляется как объекты).

У функции есть две фазы существования: место, где она определяется в программе, и место, где она вызывается.

Определять функции можно где угодно, но принято делать это в самом начале по конвенциям.

Шаблон определения функции:

    def funcname(*args, **kwargs):
      <тело функции>
      return <some values>

Что здесь что?

def - это волшебное слово, сообщающее питону, что мы хотим определить функцию.

funcname - имя нашей функции, которое мы придумываем сами. Принято писать имена функций строчными буквами. Это такая же переменная!

*args - означает неопределенное количество аргументов.

**kwargs - означает неопределенное количество аргументов, переданных по ключу (то есть, штуки наподобие sep & end у print - это аргументы, переданные по ключу).

В теле функции могут быть любые команды, ветвления, циклы.

return - оператор возвращения; его может не быть, тогда функция возвращает None. Если написать просто return без всего, тоже вернется None. Если в return прописать какой-то объект, то наша функция будет его возвращать, и его можно будет складывать в переменную или как-то иначе с ним потом работать.

```python
def area(a, h):
  """Функция, которая возвращает площадь прямоугольного треугольника"""
  return 1/2 * a * h

a, h = (int(x) for x in input('Введите основание и высоту через пробел: ').split())
s = area(a, h)
print('Площадь равна', s)
```

В этом примере я нарочно сложным образом считываю переменные a и h. Предлагаю вам попробовать самостоятельно разобраться, что там происходит такое (станет понятнее, когда пройдем списки).

Мы уже несколько раз упоминали аргументы и параметры. На самом деле, если подходить к терминологии строго, то параметры - это то, что мы пишем в скобочках функции во время ее определения, а аргументы - это конкретные объекты, которые мы ей передаем во время вызова, но обычно все это забывают, включая разрабов. 

Параметры в функции можно задавать как
- обязательные
- необязательные

Обязательные параметры необходимо будет передать при вызове функции, а необязательные имеют какие-то дефолтные значения. 

```python
def myfunc(arg1, arg2=4):
  print(arg1, arg2)

myfunc(10, 5)
myfunc(10)
```

Эти разновидности относятся только к определению функции!

С другой стороны, передавать аргументы в функцию во время **вызова** можно тоже двумя способами:

- позиционно
- по ключу (keyword)

```python
def myfunc(arg1, arg2, arg3=3):
  print(arg1, arg2, arg3)

myfunc(1, 2, 3)
myfunc(arg2=1, arg1=2)
myfunc(arg3=10, arg2=20, arg1=25)
```

Обратите внимание, что то, как мы задаем параметры в определении функции, и то, как мы их передаем в вызове функции - не взаимосвязанные, не зависящие друг от друга вещи!

Все вышеописанное касается того, как питон работает с функциями; но есть еще способы подробнее описать свою функцию: ее работоспособность это никак не изменит, а IDE начнет подсказывать по вашим описаниям. 

```python
def myfunc(arg1: list, arg2: int=2) -> str:
  return str(arg2)

myfunc([1, 2, 3], 3)
```

Функция в ячейке ничего осмысленного не делает. Когда вы начнете набирать название своей функции в теле программы, чтобы ее вызвать, питон подскажет, что он ожидает, что аргумент arg1 должен быть списком, а arg2 - числом. -> str сообщает и нам, и питону, что функция должна возвращать str (хотя не обязана), и он будет подсказывать нам, если мы попытаемся применить какой-нибудь не строчный метод к объекту функции, что ататат. А в IDE, если вы укажете -> str и напишете при этом return 0, среда разработки может выделить ваше 0 и написать, что ожидается str. 

Также можно использовать оператор pass, который ничего не делает и служит филлером: если хотите написать функцию, но тело функции еще не придумали, можно написать pass и дописать потом.

Ну и самое сложное, что связано с функциями - это понятие области видимости переменных. 

В питоне различаются глобальные и локальные переменные. Обычные переменные в теле программы - глобальные, они существуют во все время работы программы. Переменные в теле функции - локальные и существуют только во время работы функции. 

    def func(arg):
      arg = 3
      print(arg)
      arg *= 8
      print(arg)

    x = 5
    func(x)
    print(x)
    print(arg)

в такой бессмысленной программе переменная x будет существовать все время и в любой момент с ней можно что-то делать; переменная arg будет существовать только в момент, когда мы делаем func(x) - при этом питон неявно выполняет arg = x - и вызов print(arg) заставит питон заругаться, что он якобы не знает такой переменной. 

А что будет, если у нас есть глобальная и локальная переменные с одинаковым именем?

```python
def func(x):
  x += 1
  print(x)

x = 5
func(x)
print('Global x is', x)
```

Ответ простой. Собственно, все эти заморочки с областями видимости переменных как раз нужны для того, чтобы наши одноименные переменные друг другу не мешали. В момент, когда мы вызываем функцию, у нас появляется локальная переменная х, которая "затеняет" глобальную, а глобальная на время становится недоступной. Когда функция отрабатывает, глобальная х снова возвращается на сцену, потому что локальная перестала существовать. 

#### Рекурсия

Рекурсию легко понять, но непросто использовать. 

Рекурсия в программировании - это когда функция вызывает внутри себя саму себя. Так можно! Но осторожно. Если не предусмотреть точку, в которой рекурсия закончится и начнет разворачиваться назад, можно попасть в тот самый stack overflow. 

Чаще всего рекурсию показывают на факториале (факториал n - это произведение всех натуральных чисел от 1 до n)

```python
def fact(n):
  if not n:
    return 1
  return n * fact(n - 1)

fact(4)
```

Что здесь происходит? Мы передаем в функцию fact четверку. 4 != 0, а значит, not n не выполняется (осмыслите, почему). Программа переходит к строчке 5, где ей надо выполнить return и вернуть какое-то выражение. ОК, n она знает: это 4. Но потом ей нужно вычислить fact(3)! Значит, программа запоминает место, на котором она остановилась, и уходит в функцию fact(3). Снова то же самое: 3 != 0, переходим в строчку 5... n мы знаем, но нужно вычислить fact(2). Это повторяется до тех пор, пока n не станет равно 0: тогда программа с облегчением обнаружит, что нужно вернуть просто один, и вернет ее на то место, где у нее было 1 * fact(0), а там все свернется в обратную сторону, пока к нашей четверке не вернется 3!. 

#### Лямбда-функции

Допустим, у нас есть функция-крошка:

```python
def microfunc(x):
  return x ** 3
```

Эта функция ничего не делает, а только возвращает куб от своего аргумента. Сколько мишуры! Нужно написать def, придумать название функции, поставить двоеточие, написать оператор return. 

В питоне для таких микрофункций есть упрощенный синтаксис, который позволяет задавать их в одну строку. 

```python
microfunc = lambda x: x ** 3
```

Волшебное слово lambda говорит питону, что мы сейчас создадим объект типа "функция", а название ему придумывать не хотим (я в ячейке выше присвоила объект в переменную microfunc, и теперь этот объект можно вызывать как microfunc(), но это было совсем не обязательно). 

Лямбдой можно задать только такую функцию, у которой в теле нет ничего, кроме return. Эти функции удобно использовать в ключе сортировки и во многих других местах. Можно их задавать так, как я сделала в ячейке выше, это просто укоротит запись. 

```python
A = [123, 456, 33, 1234]
A.sort(key=lambda x: x % 10)
print(*A)
```

Такая сортировка позволит нам отсортировать числа не обычным способом, а по последней цифре (потому что x % 10 и возвращает последнюю цифру многозначного числа).
